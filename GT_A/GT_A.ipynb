# Gradient Tracking A

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

from utils import build_W, QuadraticDataset, LogisticDataset
```

```python
path_results = "./results"
if not os.path.exists(path_results):
    os.makedirs(path_results)
```

## Parameters

```python
# Agent parameters
m = 12 # Number of agents
n = 10 # Number of samples per agent
d = 2 # Dimension of the problem

# Function parameters
L = 1.
mu = 0.1

# Algorithm parameters
n_iter = 30000
gamma = 1e-1

# Gossip matrix
W = build_W(m, type = "Four")
```

## Quadratic Dataset
$$
f_k(\theta)
= \frac{1}{2n} \sum_{i = 1}^n \| A_{k, i}^{1/2} (\theta - \theta_{k, i}^*) \|^2.
$$

```python
dataset = QuadraticDataset(m = m, n = n, d = d, gamma = gamma, W = W, n_iter = n_iter, L = L, mu = mu)
dataset.run()
```

```python
dataset.plot_distance(idx_end = 1000, legend = False, save = os.path.join(path_results, "GT_A_Quad"))
```

```python
dataset.plot_2D(legend = False, save = os.path.join(path_results, "GT_A_Quad_Trajectory"))
```

```python
dataset.plot_legend_only(path = os.path.join(path_results, "GT_A_Legend_Trajectory"))
```

```python
gamma = 1e-1
list_W = [build_W(m, type = "Four", strength = 0.2),
          0.8 * build_W(m, type = "Four", strength = 0.9) + 0.2 * build_W(m, type = "FL"),
          0.5 * build_W(m, type = "Four", strength = 0.9) + 0.5 * build_W(m, type = "FL"),
          0.2 * build_W(m, type = "Four", strength = 0.9) + 0.8 * build_W(m, type = "FL"),
          build_W(m, type = "FL")]
colors = ["orange", "blue", "green", "red", "purple"]

n_iter = 1000

dataset = QuadraticDataset(m = m, n = n, d = d, gamma = gamma, W = list_W[0], n_iter = n_iter, L = L, mu = mu)
theta_star = dataset.theta_star.copy()
Theta_star = np.tile(theta_star, (m, 1))

result = np.zeros((len(list_W), n_iter + 1))
handles = []

for i, W_ in enumerate(list_W):
    dataset.set_W(W_)
    dataset.run()

    history = dataset.history.copy()
    result[i, :] = ((history[:, 0] - Theta_star[None, :, :])**2).sum(axis = (1, 2))

plt.figure(figsize = (8, 6))

for i, W_ in enumerate(list_W):
    eigenvalue = np.sort(abs(np.linalg.eigvals(W_)))[-2]
    plt.plot(result[i, :], label = r"$\lambda_2(W) = %.2f$" % eigenvalue, color = colors[i])
    plt.yscale('log')
    handles.append(Line2D([0], [0], label=r"$\lambda_2(W) = %.2f$" % eigenvalue, color=colors[i]))
plt.xlabel(r'Iteration $t$')
plt.ylabel(r"$\| \Theta_t - \Theta^* \|^2$")
plt.title("Convergence to the limit point - GT - Quadratic")
plt.grid()

plt.savefig(os.path.join(path_results, "GT_A_eigenvalues_Quad.pdf"), dpi=200)
plt.savefig(os.path.join(path_results, "GT_A_eigenvalues_Quad.png"), dpi=200)

plt.show()

# Plotting the legend for eigenvalues

plt.figure(figsize=(7, 0.5))
plt.legend(handles=handles, loc='center', frameon=False, ncol=len(handles))
plt.axis('off')
plt.tight_layout()
plt.savefig(os.path.join(path_results, "GT_A_eigenvalues_Legend.pdf"), dpi=200)
```

## Logistic Dataset

Here, we have
$$
f_k(\theta)
= \frac{1}{2n} \sum_{i = 1}^n \log(1 + \exp(\langle\theta, \theta_{k, i}^* \rangle)) + \frac{\mu}{2} \| \theta \|^2.
$$

```python
dataset = LogisticDataset(m = m, n = n, d = d, gamma = gamma, W = W, n_iter = n_iter, L = L, mu = mu)
dataset.run()
```

```python
dataset.plot_distance(idx_end = 1000, legend = False, save = os.path.join(path_results, "GT_A_Logi"))
```

```python
dataset.plot_2D(legend = False, save = os.path.join(path_results, "GT_A_Logi_Trajectory"), delta = .5)
```
